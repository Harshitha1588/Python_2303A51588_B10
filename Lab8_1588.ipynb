```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Step 1: Load Dataset
df = pd.read_csv('/content/population.csv')

# Step 2: Sort by population and select top 5
top5 = df.sort_values(by='Population', ascending=False).head(5)

# Step 3: Plot bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(top5['Country'], top5['Population'], color='skyblue')

# Step 4: Format y-axis to show in billions
def format_billions(x, pos):
    return f'{x * 1e-9:.1f}B'

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(format_billions))

# Step 5: Add title and labels
plt.title('Top 5 Most Populous Countries (World Population Dataset)')
plt.xlabel('Country')
plt.ylabel('Population (Billions)')

# Step 6: Add population labels above bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height,
             f'{height/1e9:.2f}B', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load Dataset
df = pd.read_csv("/content/movies_genre.csv")

# Step 2: Plot Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Genre'], df['Movie_Count'], color='skyblue', edgecolor='black')

# Step 3: Add labels and title
plt.title("Number of Movies Produced by Genres (IMDB Dataset)", fontsize=14)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)

# Step 4: Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right')

# Step 5: Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height,
             f'{height}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
